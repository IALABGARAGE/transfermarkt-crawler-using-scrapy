```python
import pandas as pd
import numpy as np

# df_europe = pd.read_csv('europe.csv')
# df_africa = pd.read_csv('africa.csv')
# df_america = pd.read_csv('america.csv')
# df_asia = pd.read_csv('asie.csv')


# df = pd.concat([df_europe, df_africa, df_america, df_asia],axis=0)
df = pd.read_csv('football_dataset.csv')
df = df.dropna(thresh=4)

# Replace element

df = df.replace(['[]','-'], np.NaN)

#Drop Nan

df = df.dropna(subset=['price'])


def delextrachar(liste):
    
        """
        Delete extra char
        
        """
        
        result = []
        
        for i,srt in enumerate(liste):
            
        
            delr = str(srt).replace("\r", "")
            delt = delr.replace("\t", "")
            deln = delt.replace("\n", "")
            delv = deln.replace(",", "")
            delvspace = delt.replace(" ", "")
            delvcrcg = delvspace.replace("[", "")   
            delvcrcd = delvcrcg.replace(" ", "")
            delvgui = delvcrcg.replace("'", "")
            final = delvgui.replace("]", "")
            result.append(final)
        
        return result
    
def virguleToPoint(word):
    delr = word.replace(",",".")
    return delr


def convert_string_to_list(srt):
    """
    Prends en entrée une string 
    Exemple : Ligue1, Coupe de la Ligue, Coupe de France
    Puis la convertit en liste 
    Exemple : [Ligue1, Coupe de la Ligue, Coupe de France]
    -----------
    Return : une liste de liste
    """
    for i, element in enumerate(srt):
        try:
            srt[i] = element.split(',')
        except:
            pass
    
    return srt   

poste_1 = delextrachar(df['poste1'].values)
poste_2 = delextrachar(df['poste2'].values)

df
```

### Nettoyage de la colonne poste 

```python
#Clean poste

liste_poste = []

all_post = ('AttackingMidfield', 'DefensiveMidfield', 'Centre-Forward', 'Centre-Back', 'Left-Back', 'RightWinger',
'LeftWinger', 'Right-Back', 'Goalkeeper','CentralMidfield', 'SecondStriker','Defender','LeftMidfield','Midfielder','RightMidfield',
'Forward','RightMidfield','Central Midfield','Attacking Midfield','Defensive Midfield','Midfielder','Right-Back','Left Winger')


for i, (poste1, poste2) in enumerate(zip(poste_1, poste_2)):
    
        if poste1 in all_post:
            liste_poste.append(poste1)
        elif poste2 in all_post :
            liste_poste.append(poste2)
        else:
            liste_poste.append(np.NaN)

            
df_poste = pd.DataFrame(liste_poste)
df_poste.columns = ['poste']
df = df.reset_index(drop=True)
df = df.drop(['poste1','poste2'],axis=1).reset_index(drop=True)
df = pd.concat([df, df_poste],axis=1)
```

```python
colrangeprice = [item.replace('Th.€', '1000') for item in df['price_range']]
colrangeprice = [item.replace('Mill.€', '1000000') for item in colrangeprice]
colrangeprice = [float(i) for i in colrangeprice]

df['price'] = df['price'].apply(lambda x: virguleToPoint(x))

colprice = [float(i) for i in df['price'].values]

price = np.multiply(colrangeprice, colprice)
df_price = pd.DataFrame(price)
df_price.columns = ['price']
df = df.reset_index(drop=True)
df = df.drop(['price', 'price_range'], axis=1)
df = pd.concat([df, df_price], axis=1)
df = df.dropna(subset=['poste'])
df.head(5)
```

### Nettoyage de l'age

```python
def get_age(age):

    liste = []
    for i, element in enumerate(age):
        try :
            liste.append(int(element[element.find("(")+1:element.find(")")]))
        except:
            pass
    return liste

age_clean = get_age(df['age'])
dfage = pd.DataFrame(age_clean)
dfage.columns = ['age']
df = df.dropna(subset=['age'])
df = df.reset_index(drop=True)
df['age'] = dfage['age']
```

### Nettoyage fin de contrat 

```python
def clean_fin_contrat(data):
    
    current_date = 2018
    result = []
    for i, element in enumerate(data):
        try:
            result.append(int(element.split('.')[2]) - current_date)
        except:
            result.append(np.NaN)
        
    return result

dfcontrat = pd.DataFrame(clean_fin_contrat(df['fin_contrat'].dropna()))
dfcontrat.columns = ['fin_contrat']
df = df.dropna(subset=['fin_contrat'])
df = df.reset_index(drop=True)

df = df.drop(['fin_contrat'],axis=1)
df = pd.concat([df, dfcontrat],axis=1)
```

```python
def generate_stats(stat):
    """
    Convertit une liste de (4*Nombre de competition jouées + 1) en un tableau de données de
    dimension (Nombre de competition jouées + 1) X 4
    ------------------
    Return : une liste de stat par compétition
    """
    liste_of_stats = []
    
    try:
        for j in range(10,len(stat), 9):
            liste_of_stats.append(stat[j:j+9])
    except:
        return np.zeros(9)
    
    return liste_of_stats


generate_stats(stats[14])
```

```python
def clean_stat(listofstats):

    """
    Convertit en float les données + remplace les tirets 
    par des 0

    """
    for i, stat in enumerate(listofstats):
        stat = [item.replace('-', '0') for item in stat]
        try:
            listofstats[i] = list(map(float, stat))
        except:
            pass
        
    return listofstats

```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.07, random_state=42)

from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras import optimizers


# Initialising the ANN : création des différentes couches

model = Sequential()

model.add(Dense(units=256, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=256, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(1))


opt = optimizers.Adam(lr=0.00005, decay=1e-10)

model.compile(optimizer=opt, loss='mse', metrics=['mae'])

callbacks = [EarlyStopping(monitor='loss', patience=12)]


history = model.fit(X_train, y_train,
                             batch_size=1,
                             epochs=500,
                             callbacks=callbacks,
                             validation_data=(X_test, y_test))


import matplotlib.pyplot as plt

print(history.history.keys())

# summarize history for loss
plt.plot(history.history['mean_absolute_error'])
plt.plot(history.history['val_mean_absolute_error'])

plt.title('model mean_absolute_error')
plt.ylabel('mean_absolute_error')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

print(pd.concat([pd.DataFrame(model.predict(X_test)),y_test.reset_index(drop=True)], axis=1))
```
